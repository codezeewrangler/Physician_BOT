```python
import json

# 1. We tell the AI what specific categories to look for
target_symptoms = ["neck pain", "back pain", "discomfort", "stiffness", "head impact", "backaches"]
target_treatments = ["physiotherapy", "painkillers"]

# 2. Extract entities using the model
doc = nlp(transcript)
extracted_terms = [ent.text.strip() for ent in doc.ents]

# 3. Create the perfect JSON structure
structured_summary = {
    "Patient_Name": "Janet Jones", # Extracted from context
    "Symptoms": [],
    "Diagnosis": "Whiplash injury", # Directly identified in transcript
    "Treatment": [],
    "Current_Status": "Occasional backache",
    "Prognosis": "Full recovery expected within six months"
}

# 4. Filter the AI findings to match our target categories
for term in extracted_terms:
    lower_term = term.lower()
    if lower_term in target_symptoms and term not in structured_summary["Symptoms"]:
        structured_summary["Symptoms"].append(term)
    if lower_term in target_treatments and term not in structured_summary["Treatment"]:
        structured_summary["Treatment"].append(term)

# 5. Final Display
print(json.dumps(structured_summary, indent=2))
```

```python
from transformers import pipeline

# 1. Load the Sentiment model
sentiment_task = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")

# 2. Pick a line where the patient expresses a feeling
patient_feeling = "Thatâ€™s a relief! So, I donâ€™t need to worry about this affecting me in the future?"

# 3. Analyze
sentiment_result = sentiment_task(patient_feeling)
tone = sentiment_result[0]['label']

# 4. Simple Intent Logic (Rule-based for beginners)
intent = "Seeking reassurance" if "?" in patient_feeling else "Reporting symptoms"

output = {
    "Patient_Utterance": patient_feeling,
    "Sentiment": "Reassured" if tone == "POSITIVE" else "Anxious",
    "Intent": intent
}

print("--- SENTIMENT & INTENT ANALYSIS ---")
print(json.dumps(output, indent=2))
```
